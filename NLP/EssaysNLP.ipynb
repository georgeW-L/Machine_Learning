```python
import pandas as pd
from dateutil.parser import parse
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv('/content/drive/MyDrive/Machine Learning 23 24/Data/output_dataframe.csv')
```

```python
df
```

```python
df.isna().sum()
```

```python
df_new = df.dropna()
```

```python
df_new
```

```python
def remove_dates(text):
    date_pattern = re.compile(r'\b\d{1,4}[-/]\d{1,2}[-/]\d{1,4}\b') #compiles regular expression pattern into regular expression object, in other words creates a pattern to search for
    return re.sub(date_pattern, '', text) if text else None #re.sub() replaces pattern with specified text

# Apply the function to the 'Text Content' column using apply
df_new['text_without_dates'] = df_new['Text Content'].apply(remove_dates)
```

```python
df_new
```

```python
X = df_new['Document Name'] + ' ' + df_new['Author Name'] + ' ' + df_new['text_without_dates'] #TF-IDF vectorization expects iterable objects not dataframes
y = df_new['Class']

# Label encoding for the target variable
y = pd.factorize(y)[0]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.7, random_state=42)

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Support Vector Machine Classifier
svm_clf = SVC(kernel='linear')  # better results with linear kernal than sigmoid, didn't test others like rbf it ran forever
svm_clf.fit(X_train_tfidf, y_train)

# Predictions
y_pred = svm_clf.predict(X_test_tfidf)

# Model Evaluation
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2f}')

# Additional metrics
print(classification_report(y_test, y_pred))
```
