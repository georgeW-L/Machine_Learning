```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

from nltk.corpus import stopwords

from nltk.tokenize import word_tokenize

from nltk.stem import WordNetLemmatizer
```

```python
print(stopwords.words('english'))
```

```python
def preprocess_text(text):

    # Tokenize the text

    tokens = word_tokenize(text.lower())




    # Remove stop words
    # List comprehension that removes tokens that are prepositions from data
    filtered_tokens = [token for token in tokens if token not in stopwords.words('english')]




    # Lemmatize the tokens
    # process by which words are reduced to their base form
    lemmatizer = WordNetLemmatizer()

    lemmatized_tokens = [lemmatizer.lemmatize(token) for token in filtered_tokens]




    # Join the tokens back into a string

    processed_text = ' '.join(lemmatized_tokens)

    return processed_text
```

```python
df['OriginalTweet'] = df['OriginalTweet'].apply(preprocess_text)
```

```python
df
```

```python
# initialize NLTK sentiment analyzer

#analyzes text and returns scores for negativity, neutrality, positivity and a compound score
analyzer = SentimentIntensityAnalyzer()

def get_sentiment(text):

    scores = analyzer.polarity_scores(text)
    if scores['compound'] > .75:
      sentiment = 1
    elif scores['compound'] > .25:
      sentiment = .5
    elif scores['compound'] > -.25:
      sentiment = 0
    elif scores['compound'] > -.75:
      sentiment = -.5
    else:
      sentiment = -1
    return sentiment



df['sentiment'] = df['OriginalTweet'].apply(get_sentiment)
```

```python
df['sentiment'] = df['sentiment'].astype(str)
```

```python
from sklearn.metrics import classification_report

print(classification_report(df['Sentiment'], df['sentiment']))
```
